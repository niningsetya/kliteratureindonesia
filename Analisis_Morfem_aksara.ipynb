```python
# "rawdata.txt" 파일의 내용을 "원본데이터" 변수로 불러오기
import pandas as pd
원본데이터 = pd.read_csv('rawdata.csv', engine='python' )
원본데이터
```

```python
원본데이터['content'] = 원본데이터['content'].str.replace("/","") # / 기호 삭제
#원본데이터['content'] = 원본데이터['content'].str.replace('\r', ' ') # 줄바꿈 삭제
원본데이터['content'] = 원본데이터['content'].str.replace('\s+', ' ') # 2번 이상 띄어쓰기 삭제
원본데이터['content'] = 원본데이터['content'].str.replace('@[\w_]+', '') # enticons 제거
원본데이터['content'] = 원본데이터['content'].str.replace('[#]+[0-9a-zA-Z_]+', ' ') # Hashtag 제거
원본데이터['content'] = 원본데이터['content'].str.replace('[^a-zA-Z]', ' ') # 영어제거
#원본데이터['content'] = 원본데이터['content'].str.replace('😂', ' ') # Hashtag 제거

원본데이터['content'] = 원본데이터['content'].str.replace('RT @[\w_]+: ', '') # Retweets 제거
원본데이터['content'] = 원본데이터['content'].str.replace('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', ' ') # http로 시작되는 url 제거
원본데이터['content'] = 원본데이터['content'].str.replace('[-a-zA-Z0-9@:%._\+~#=]{1,256}\.[a-zA-Z0-9()]{2,6}\b([-a-zA-Z0-9()@:%_\+.~#?&//=]*)', ' ') # http로 시작되는 않는 url 제거

    #     pattern = '(http|ftp|https)://(?:[-\w.]|(?:\da-fa-F]{2}))+'
    #     text = re.sub(pattern = pattern, repl = ' ',string=text)


원본데이터
```

```python
# "원본데이터"를 대상으로 "aksara"를 사용해서 형태소 분석하기
from aksara.pos_tagger import POSTagger
pos_tagger = POSTagger()

저장공간  = []
for index, row in 원본데이터.iterrows():# 원본데이터의 컬럼(열)을 불러와서 반복할 준비!
    대상텍스트 = row[6] # 분석에서 사용할 텍스트 정보가 있는 열을 지정해준다. 주의! python은 0부터 시작한다!
    분류 = row[5] # 분석에서 사용할 분류 정보가 있는 열을 지정해준다. 주의! python은 0부터 시작한다!
    형태소분석결과= pos_tagger.tag(대상텍스트)  # aksara pos 기능을 활용해서 대상 텍스트를 형태소 분리한다.
#    print(분류)
    for token in 형태소분석결과:
      for (형태소, 품사) in token:
        형태소종합 = [형태소,품사,분류,1]  # 형태소종합에 분류정보와 카운트정보를 추가한다.
        저장공간.append(형태소종합) # 형태소종합의 내용을 종합하여 저장한다.
```

```python
# "저장공간" 변수의 내용을 데이터프레임 형식으로 변환해서 "분석통합" 변수에 저장한다.
분석통합 = pd.DataFrame(저장공간)
분석통합
```

```python
# "분석통합" 변수의 컬럼(열) 이름 변경
분석통합.columns = ["형태소", "품사", "분류", "카운트"]
# "분석통합" 변수의 저장 내용 확인
분석통합
```

```python
# "분석통합" 품사 중 "_" 삭제
# "X" 도 삭제 필요?

분석통합 = 분석통합[분석통합.품사 != '_']
분석통합
```

```python
# "분석통합" 변수의 내용을 형태소, 품사, 분류가 같은 것을 합치고, 카운트의 총합을 구해서 "그룹통합" 변수에 저장한다.
#그룹통합 = 분석통합.groupby(['형태소', '품사', "분류"])['카운트'].sum().reset_index()
그룹통합 = 분석통합.groupby(['형태소', '품사', '분류'])['카운트'].sum().reset_index()

# "그룹통합" 변수의 저장 내용 확인
그룹통합
```

```python
# "그룹통합" 변수의 내용을 "형태소분석결과.csv" 파일로 저장한다.
# header는 컬럼(열) 정보의 포함 여부이다.
# encoding은 문자코드를 선택하는 것이다. python에서는 기본적으로 utf-8(유니코드)를 사용한다.

#그룹통합.to_csv('형태소분석결과.csv', header='true', encoding='utf-8')

# 엑셀 파일로 저장하는 방법
그룹통합.to_excel('형태소분석결과.xlsx')
```
